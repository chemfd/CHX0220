The relative affinity of binding of a ligand, $L$, to a receptor $R$, to form the complex $[R \bullet L]$ in the presence of an inhibitor (which forms a complex with the receptor $R \bullet I$ is given by:

<p align="center">

$\frac{[R \bullet L]_I}{[R \bullet L]_0}= \frac{K_d + [L]}{K_d \left(1+\frac{[I]}{K_{d,I}} \right)+[L]}$
    
</p>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


# Figure drawing for inhibition of binding:



Kdisinhibitor = 10
Kdis = 1
concinhibitor = 10
concligand = np.linspace(0,100,101)

ratiocomplexinhibitor = (Kdis+concligand)/(Kdis*(1+(concinhibitor/Kdisinhibitor))+concligand)
                                           
                                           
concligandKb = Kdis*concligand
                                           
sns.set_context('talk') #fancy very quick way to set how the graph looks using seaborn

fig,ax1 = plt.subplots(figsize=(3,5)) #setting the size
plt.title('The effect of inhibitor on ligand binding') #my title doh!
ax1.plot(concligandKb, ratiocomplexinhibitor, "o", color='#7570b3') #the data points - just choosing colours which should be good for the colourblind
ax1.set_ylabel(r'$[R \bullet L]_i / [R \bullet L]_0$') #labelling my axis - I can't remember what the r was for...
ax1.set_xlabel(r'$[L]K_b$')                                           

plt.savefig('bindinginhibition',transparent=True)
```
